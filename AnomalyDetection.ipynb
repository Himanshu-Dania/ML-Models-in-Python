```python
import numpy as np
```

```python
class Model:
    def __init__(self,X,Xval,yval):
        self.X=X
        self.Xval=Xval
#         self.y=y
        self.yval=yval
        self.m,self.n=X.shape
        self.mv,self.nv=Xval.shape
    
    def calcGaussian(self):
        self.mu= np.sum(self.X,axis=0)/self.m
        self.var=np.sum(np.square(self.X-self.mu), axis=0)/(self.m-1)
        
    def calcProb(self):
        
        self.var=np.diag(self.var)
        
        p= (2*np.pi)**(-self.n/2)*np.linalg.det(self.var)**(-1/2)*np.exp(-0.5*np.sum((self.Xval-self.mu) @ np.linalg.pinv(self.var) * ((self.Xval-self.mu)),axis=1))
        self.p=p
        print(p)
    
    def estimateThreshold(self):
        for e in np.arange(np.min(self.yval),np.max(self.yval),0.01):
            pred= (self.yval>=e)
            bestF1=0
            print(pred)
            tp,fp,fn=0,0,0
            for i in range(self.mv):
                if (self.yval[i]==1 & pred[i]==True):
                    tp=tp+1
                elif(self.yval[i]==0 & pred[i]==True):
                    fp=fp+1
                elif (self.yval[i]==1 & pred[i]==False):
                    fn=fn+1
            p=tp/(tp+fp)
            r=tp/(tp+fn)
            f1= 2*p*r/(p+r)
            if(f1>bestF1):
                bestF1=f1
                self.beste=e  
        print(self.beste)
```

```python
X=np.array([[1,5],[2,4],[3.5,3.5],[9,0],[1,6]])
Xval=np.array([[1,5.4],[3,4.2],[2,3.5],[0,0],[9,6]])
yval=np.array([0,0,0,1,1])
m=Model(X,Xval,yval)
m.calcGaussian()
m.mu,m.var
m.calcProb()
m.estimateThreshold()
# np.array([[1],[2]]).shape
# np.sum(X,axis=0)
```
