```python
import numpy as np
```

```python
class Model:
    def __init__(self, y ,r, n, lr,reg,j ):
        self.y=y
        self.r=r
        self.n=n
        self.m,self.u=self.y.shape
        self.lr=lr
        self.reg=reg
        self.j=j
        
    def initialize(self):
        self.theta=np.random.rand(self.u,self.n)
        self.X=np.random.rand(self.m,self.n)
        
    def cost(self):
        self.J= np.sum(np.sum(np.square(((self.X @ self.theta.T)-self.y)*self.r)))/2
        self.J+=self.reg/2 * (np.sum(np.sum(np.square(self.X)))+np.sum(np.sum(np.square(self.theta))))
        
        self.xgrad=((self.X@self.theta.T - self.y)*self.r)@self.theta + self.reg*self.X
        self.thetagrad= ((self.X@self.theta.T - self.y)*self.r).T @ self.X + self.reg*self.theta
        
    def graddescent(self):
        for i in range(self.j):
            self.cost()
            self.X=self.X-self.lr/self.m * self.xgrad
            self.theta=self.theta-self.lr/self.m * self.thetagrad
            print(f"Iteration {i}: Cost {self.J}\n")
            if(i%9==0):
                print(f"X={self.X}\n\n t= {self.theta}\n\n, pred={self.X@self.theta.T}\n\n")
            
            
```

```python
y=np.array([[5,0 , 5],
            [4, 1, 0],
            [2, 5 ,0],
            [0, 3, 3]])
r=np.array([[1,0 , 1],
            [1, 1, 0],
            [1, 1 ,0],
            [0, 1, 1]])
m=Model(y,r, 3,0.1, 0.01, 100)
m.initialize()

m.graddescent()
```
