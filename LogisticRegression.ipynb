```python
import numpy as np
```

```python
class Model:
    def __init__(self,X, y, theta, lr, niter, reg):
        self.X=X
        self.m,self.n=X.shape
        self.y=y
        self.theta=theta
        self.lr=lr
        self.niter=niter
        self.reg=reg
    
    def hyp(self):
        return 1/(1+np.exp(np.dot(-1* self.X,self.theta)))
    
    def cost(self):
        h=self.hyp()
        return 1/(self.m) * ( sum(-1*self.y* np.log(h) -1*(1-self.y)* np.log(1-h)) + self.reg/2*np.sum(np.square(self.theta)))
    
    def gradient_descent(self):
        for i in range(self.niter):
            cost=self.cost()
            self.theta=self.theta - self.lr/self.m * ( np.dot((self.X.T), self.hyp()-self.y) + self.reg*self.theta)
            if i%50==0:
                print(f"iteration {i}:\n{self.theta} has cost {cost}")
        return self.theta
    def train(self):
        return self.gradient_descent()
```

```python
a=np.array([[1,2],[1,4], [1,3], [1,9],[1,-5]])
b=np.array([[0],[1],[1],[1],[0]])
t=np.array([[-3],[1]])
m=Model(a,b,t, 0.1, 10000,0)
# # m.hyp()
tf= m.train()
print(f"{tf}")
H=1/(1+np.exp(np.dot(-1*a, t)))
# 1/(5) * ( sum(-1*b* np.log(h) -1*(1-b)* np.log(1-h)))
# 1/5 * ( np.dot((a.T),H -b))
```
