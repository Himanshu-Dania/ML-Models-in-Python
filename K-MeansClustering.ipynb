```python
import numpy as np
import random as r
```

```python
class Model:
    def __init__(self,X, k,niter,cent):
        self.X=X
        self.m,self.n=X.shape
        self.idx=np.zeros([self.m,1])
        self.k=k
        self.niter=niter
        self.centroids=cent
        
    def ClusterInit(self):
        self.centroids=np.zeros([self.k,self.n])
        for i in range(self.k):
            self.centroids[i]=self.X[r.randint(0,self.m-1)]
    def AssignToCluster(self):
        dist=np.zeros([self.m,self.k])
        for i in range(self.k):
            dist[:,i]=np.sum(np.square(self.X-self.centroids[i,:]), axis=1).T
        self.idx=np.reshape(np.argmin(dist, axis=1) , [self.m,1])
        return self.idx
    def NewPosition(self):
        nc=np.zeros([self.k,self.n])
        
        for i in range(self.k):
            x=0
            for j in range(self.m):
                if self.idx[j]==i :
                    x+=1
                    nc[i,:]+=self.X[j,:]
            if x!=0:
                nc[i,:]=nc[i,:]/x            
        self.centroids=nc
#         print(f"{self.centroids}")
            
       
    
    def train(self):
#         self.ClusterInit()
        print(f"{self.centroids}")
        for i in range(self.niter):
            print(f"Iteration {i}/100: {self.centroids}")
            print(self.AssignToCluster())
            self.NewPosition()
        return self.centroids, self.idx   
```
